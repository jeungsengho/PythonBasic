```python
score = int(input('점수를 입력하세요 >>> '))

if score >= 90:
  hak = 'A'
elif score >= 80:
  hak = 'B'
elif score >= 70:
  hak = 'C'
elif score >= 60:
  hak = 'D'
else:
  hak = 'F'
  
print(f'점수는 {score}이고, 학점은 {hak} 학점 입니다.')
```

```python
no = int(input('정수를 입력하세요 >>>'))

if no % 3 == 0:
  print(f'{no}는 3의 배수입니다.')
else:
  print(f'{no}는 3의 배수가 아닙니다.')
```

```python
no1 = int(input('정수 1 입력 >>> '))
no2 = int(input('정수 2 입력 >>> '))
no3 = int(input('정수 3 입력 >>> '))

if no1 > no2 and no1 > no3:
  max = no1
elif no2 > no1 and no2 > no3:
  max = no2
else:
  max = no3

print(f'가장 큰 수는 {max} 입니다.')
```

```python
no = input('차량번호를 입력하세요 >>> ')

if int(no[-1]) % 2 == 0:
  print(f"차량번호 '{no}'는 오늘 운행 가능입니다.")
else:
  print(f"차량번호 '{no}'는 오늘 운행 불가입니다.")
```

```python
# 반복문
# while   반복횟수가 불명확할 때 주로 사용
# for     반복횟수가 명확할 때 주로 사용


          # while 조건식:
          #   반복실행문

while True:  # 무한반복
  print('Hello Python')
```

```python
n = 1

while n <= 10:
  print(n)
  n += 1
```

```python
n = 10

while n >= 1:
  print(n)
  n -= 1
```

```python
day = 1
while day <= 5:
  hour = 1
  while hour <= 3:
    print(f'{day}일차 {hour}입니다.')
    hour += 1
  day += 1
```

```python
no = int(input('정수를 입력하세요 >>> '))

if no <= 0:
  print('잘못된 입력입니다.')
else:
  n = 0
  while no > n:
    print(f'{n+1}번째 Hello')
    n += 1
```
